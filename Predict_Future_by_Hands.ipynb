```python
from google. colab                 import drive
from matplotlib.image              import imread
from os                            import listdir
from numpy                         import asarray
from numpy                         import save
from keras.utils                   import load_img
from keras.utils                   import img_to_array
from keras.models                  import Sequential
from keras.layers                  import Dense, Activation, BatchNormalization, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.utils        import load_img, img_to_array
from google.colab                  import drive
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder= '/content/drive/MyDrive/AI_PROJECT/Hand/'

photos, labels = list(), list()
for file in listdir(folder):

  if file.startswith('suckhoe'):
    output = 0
    photo = load_img(folder+file, target_size=(200,200))
    photo = img_to_array(photo)
    photos.append (photo)
    labels.append(output)

  elif file.startswith('thongminh'):
    output = 1
    photo = load_img(folder+file, target_size=(200, 200) )
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  elif file.startswith('giauco'):
    output = 2
    photo = load_img(folder+file, target_size=(200, 200) )
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  elif file.startswith('chamchi'):
    output = 3
    photo = load_img(folder+file, target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  elif file.startswith('khodau'):
    output = 4
    photo = load_img(folder+file, target_size=(200, 200) )
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  elif file.startswith('dinhmenh'):
    output = 5
    photo = load_img(folder+file, target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)
  elif file.startswith('thanhdat'):
    output = 6
    photo = load_img(folder+file, target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)
  elif file.startswith('omyeu'):
    output = 7
    photo = load_img(folder+file, target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)
  elif file.startswith('cancu'):
    output = 8
    photo = load_img(folder+file,target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)
  elif file.startswith('giandoi'):
    output = 9
    photo = load_img(folder+file,target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)
photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)

save('10hands_photos.npy', photos)
save('10hands_labels.npy', labels)
```

```python
print(photos.shape)
print(labels.shape)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation= 'relu',
                 kernel_initializer= 'he_uniform',
                 padding='same', input_shape=(200,200, 3)))

model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu',
                 kernel_initializer= 'he_uniform',
                 padding='same'))

model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu',
                 kernel_initializer= 'he_uniform',
                 padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(128, (3, 3), activation='relu',
                 kernel_initializer= 'he_uniform',
                 padding='same'))

model.add(MaxPooling2D((2, 2)) )
model.add(Flatten())
model.add(Dropout(0.2))

model.add(Dense(128, activation= 'relu'))
model.add(Dense(128, activation= 'relu'))
model.add(Dense(128, activation= 'relu'))
model.add(Dense(128, activation= 'relu'))
model.add(Dense(128, activation= 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(photos, labels, batch_size=64,epochs=100, verbose=1)

model.save('10hands.h5')

classes = ["suckhoe","thongminh", "giauco","chamchi", "khodau", "dinhmenh", "thanhdat", "omyeu", "cancu","giandoi"]
```
